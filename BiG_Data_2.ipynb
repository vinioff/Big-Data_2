# Corrigir as colunas com dados textuais com One Hot Encoding ou Label Enconding

```python
# usei One Hot Encoding pois a coluna "season" tem as estações em texto,
# mais especificamente a palavra Spring que não tem sentindo nenhum em uma vizualização adequada.
# Então aqui eu transformo cada estação em uma coluna,
# se for 0 significa que não é aquela estação, senão 1 é
estacoes = pd.get_dummies(dados['season'], prefix='estação', dtype=int)
```

```python
estacoes.head()  # vendo o resultado
```

```python
dados_certos = pd.concat([dados, estacoes], axis=1)  # colando as colunas novas das estações que acabei de criar na tabela principal
```

```python
dados_certos.drop(columns=['season'] , inplace=True) # excluindo a coluna "season" original, que era textual e não agrega em nada na nova tabela
```

```python
dados_certos.head()
```

```python
# tratando os dados da coluna holiday, deixando o  No em 0 e Yes em 1
dados_certos['holiday'] = dados_certos['holiday'].map({'No': 0 , 'Yes': 1})

dados_certos['workingday'] = dados_certos['workingday'].map({'No': 0 , 'Yes': 1})
```

```python
# tratando os dados da coluna ano, se for 0 será igual a 2018, e se for 1 será igual a 2019
dados_certos['yr'] = dados_certos['yr'].map({0.0: 2018 , 1.0: 2019})
```

# Realizar o estudo de hiperparâmetros para 4 modelos de Machine Learning



```python
# Dividindo X e y (Entrada e saída de dados)
X = dados_certos.drop(columns=["cnt"])
y = dados_certos["cnt"]

# Dividindo dados para treino e dados para teste
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state= 42
                                                    )

scale = MinMaxScaler().fit(X_train)

# normalização no treino e no teste
X_train = scale.transform(X_train)
X_test = scale.transform(X_test)
```

```python
# GridSearch "automatizado"
# Definindo os modelos e parâmetros em um dicionário

# arvore de decisao
arv = {"modelo":DecisionTreeRegressor(),
      'param': {'max_depth': range(2,21),
                'min_samples_leaf': range(2,11)}}

# floresta aleatoria
fr = {"modelo":RandomForestRegressor(),
      'param': {'max_depth': range(2,11),
                'min_samples_leaf': range(3,10)}}

# gradient
gb = {"modelo":GradientBoostingRegressor(),
      'param': {'max_depth': range(2,11),
                'min_samples_leaf': range(3,10)}}

# Junto todos os modelos num dicionário só
grid_search = { 'AR' : arv,
                'FR': fr,
                'GB' : gb}
```

```python
best_param = {}
for model, dict_GS in grid_search.items():
  modelo_GS = GridSearchCV(dict_GS['modelo'],
                           dict_GS['param'])
  modelo_GS.fit(X_train, y_train)
  best_param[model] = modelo_GS.best_params_
```

# Treinamento de 4 modelos diferentes de Machine Learning

```python
modelos = { 'AR' : DecisionTreeRegressor(max_depth=14, min_samples_leaf=5),
            'FR' : RandomForestRegressor(max_depth=10, min_samples_leaf=3),
            'GB' : GradientBoostingRegressor(max_depth=10, min_samples_leaf=9)}

# Guardando as previsões de cada modelo
previsoes = {}
for nome_modelo, modelo in modelos.items():
  regressor = modelo
  modelo.fit(X_train,y_train)
  previsoes[nome_modelo]=modelo.predict(X_test)
```

# Levantar as métricas de desempenho de cada um dos modelos treinados

```python
# avaliando cada modelo
for nome_modelo, prev in previsoes.items():
  print('_'*40)
  print(f"{nome_modelo:^39}")

  mse = mean_squared_error(y_true=y_test,
                           y_pred=prev)
  mae = mean_absolute_error(y_true=y_test,
                            y_pred=prev)
  r2 = r2_score(y_true=y_test,
                y_pred=prev)

  # mostrando os valores
  print(f"Erro Quadrático Médio = {mse}")
  print(f"Erro Absoluto Médio   = {mae}")
  print(f"Coeficiente R²        = {r2}")
```

# Justificativa do melhor modelo


O **Gradient Boosting (GB)** foi o melhor modelo entre os testados (Árvore de Decisão, Floresta Aleatória e Gradient Boosting).

O **Gradient Boost Regressor** teve o melhor desempenho, com:
- **Menores valores de erro** (MSE e MAE) por uma margem considerável
- **Maior valor de R²**, mostrando que explicou melhor a variação dos dados

Os outros modelos funcionaram bem, mas **nenhum superou o GB nos três indicadores ao mesmo tempo**, tornando-o a melhor escolha para este caso.
